```python
plt.figure(figsize=(16,6))
sns.displot(df['Age'],bins=20,color='red')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.swarmplot(x='Attrition', y='Age',data=df,hue='Attrition')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(data=df, x='Department')
```

```python
plt.figure(figsize=(16, 10))
sns.countplot(data=df, x='EducationField')  
plt.xlabel('Education Field', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.title('Distribution of Employees by Education Field', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.swarmplot(x='Attrition',y='MonthlyIncome',data=df)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.regplot(x='JobLevel',y='MonthlyIncome',data=df)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 18), gridspec_kw={'hspace': 0.2})
fig.suptitle('Histogram for all numerical variables in the dataset')
```

```python
numeric = df.select_dtypes(include=[np.number])
plt.figure(figure=(17,10))
sns.heatmap(numeric.corr(),annot=True,cmap='rainbow')
plt.show()
```
